```python
# importing required libraries of opencv
import cv2

# importing library for plotting
from matplotlib import pyplot as plt

# reads input image
img1 = cv2.imread('night.jpg')
img2 = cv2.imread('day.jpg')
img3 = cv2.imread('high.png')
img4 = cv2.imread('low.jpg')

# find frequency of pixels in range 0-255
histr1 = cv2.calcHist([img1],[0],None,[256],[0,256])
histr2 = cv2.calcHist([img2],[0],None,[256],[0,256])
histr3 = cv2.calcHist([img3],[0],None,[256],[0,256])
histr4 = cv2.calcHist([img4],[0],None,[256],[0,256])

# for subplotting
plt.subplot(241),plt.imshow(img1,'gray'),plt.title('Low Light')
plt.subplot(242),plt.imshow(img2,'gray'),plt.title('Day Light')
plt.subplot(243),plt.imshow(img3,'gray'),plt.title('High Contrast')
plt.subplot(244),plt.imshow(img4,'gray'),plt.title('Low Contrast')
plt.subplot(245),plt.plot(histr1)
plt.subplot(246),plt.plot(histr2)
plt.subplot(247),plt.plot(histr3)
plt.subplot(248),plt.plot(histr4)
plt.show()

```
