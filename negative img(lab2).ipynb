```python
#negative and thresholding of  an image
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

#Using 0 to read image in grayscale mode
img = cv.imread("messi.jpg",0)

#The function cv.threshold is used to apply the thresholding. 
#The 1st arg is the source image, which should be a grayscale image.
#The 2nd arg is the threshold value which is used to classify the pixel values. 
#The 3rd arg is the maximum value which is assigned to pixel values exceeding the threshold. 
#THRESH_BINARY  for setting the dest. pixel to maxValue if corres. src pix > threshold pix, 
#else  to zero 
#THRESH_BINARY_INV  is just the opposite of binary thresholding.
ret,thresh1=cv.threshold(img,127,255,cv.THRESH_BINARY)
ret,thresh2=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)

titles=["Original Image","Threshold Image","Negative Image"]
images=[img,thresh1,thresh2]

for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#import required libraries of opencv
import cv2

#import library for plotting
from matplotlib import pyplot as plt

img = cv2.imread('messi.jpg',0)
img1=cv2.imread('1.jpg',0)
img2=cv2.imread('2.png')
histr=cv2.calcHist([img],[0],None,[256],[0,256])
histr1=cv2.calcHist([img1],[0],None,[256],[0,256])
histr2=cv2.calcHist([img2],[0],None,[256],[0,256])
plt.subplot(231),plt.imshow(img,'gray'),plt.title('Low Light')
plt.subplot(232),plt.imshow(img1,'gray'),plt.title('Low Contrast')
plt.subplot(233),plt.imshow(img2,'gray'),plt.title('Day light')
plt.subplot(234),plt.plot(histr)
plt.subplot(235),plt.plot(histr1)
plt.subplot(236),plt.plot(histr2)
plt.show()
```
